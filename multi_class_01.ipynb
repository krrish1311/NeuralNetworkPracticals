```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('wine.csv')
```

```python
dataset.corr()
```

```python
dataset.info()
```

```python
y =dataset["Wine"]
```

```python
X = dataset.drop("Wine" , axis=1)
```

```python
X
```

```python
y.value_counts()
```

```python
y_cat = pd.get_dummies(y)
```

```python
from keras.models import Sequential
```

```python
model  = Sequential()
```

```python
from keras.layers import Dense
```

```python
model.add(Dense(units=32 , kernel_initializer='he_normal',activation='relu'))
```

```python
model.add(Dense(units=16  , kernel_initializer = "he_normal"  ,activation='relu'))
```

```python
model.add(Dense(units=8 , kernel_initializer='he_normal',activation='relu'))
```

```python
model.add(Dense(units=3 , kernel_initializer='he_normal',activation='softmax'))
```

```python
model.get_config()
```

```python
from keras.optimizers import RMSprop
```

```python
#what is diference between 'accuracy ' and 'Accuracy
'
```

```python
model.compile(optimizer=RMSprop(learning_rate=0.001),loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
model.fit(X,y_cat, epochs=50)
```

```python
pred = model.history.history["accuracy"]
```

```python
plt.plot(pred)
```
